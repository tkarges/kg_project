```python
import json
import pandas as pd
import numpy as np

json_file_path = 'mmds_master.json'
# json_file_path = 'mk_wifo_master.txt'

# Read the JSON file
with open(json_file_path, 'r') as f:
    data = json.load(f)

# Create a pandas DataFrame from the JSON data
df_modules = pd.DataFrame(data)

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

print("DataFrame reloaded for semester cleaning. Displaying first 5 rows:")
display(df_modules.head())
```

```python
def clean_form_of_module(text):
    text_lower = str(text).lower()
    found_terms = []

    # Check for keywords and add them to the list if found
    if 'lecture' in text_lower:
        found_terms.append('Lecture')
    if 'inverted classroom' in text_lower:
        found_terms.append('Inverted classroom')
    if 'exercise' in text_lower:
        found_terms.append('Exercise')
    if 'tutorial' in text_lower:
        found_terms.append('Tutorial')
    if 'project' in text_lower:
        found_terms.append('Project')

    # Remove duplicates while preserving order
    seen = set()
    unique_found_terms = []
    for item in found_terms:
        if item not in seen:
            seen.add(item)
            unique_found_terms.append(item)

    if unique_found_terms:
        return ', '.join(unique_found_terms)
    else:
        # If no keywords are found, return the original text (lowercased) as a fallback
        return text_lower

df_modules['form_of_module'] = df_modules['form_of_module'].apply(clean_form_of_module)

print("Unique values in 'form_of_module' column after refined cleaning:")
display(df_modules['form_of_module'].value_counts())
```

```python
def clean_type_of_module(text):
    text_lower = str(text).lower()
    if 'specialization course' in text_lower or 'specialization' in text_lower:
        return 'Specialization Course'
    return text

df_modules['type_of_module'] = df_modules['type_of_module'].apply(clean_type_of_module)

print("Unique values in 'type_of_module' column after cleaning:")
display(df_modules['type_of_module'].value_counts())
```

```python
def clean_form_of_assessment(text):
    text_lower = str(text).lower()
    found_terms = []

    if 'written exam' in text_lower:
        found_terms.append('Written exam')
    if 'project report' in text_lower:
        found_terms.append('Project report')
    # Check for 'oral presentation' or just 'presentation'
    if 'presentation' in text_lower:
        found_terms.append('Oral presentation')
    if 'oral exam' in text_lower:
        found_terms.append('Oral exam')
    if 'learning portfolio' in text_lower:
        found_terms.append('Learning portfolio')

    # Remove duplicates while preserving order
    seen = set()
    unique_found_terms = []
    for item in found_terms:
        if item not in seen:
            seen.add(item)
            unique_found_terms.append(item)

    if unique_found_terms:
        return ', '.join(unique_found_terms)
    else:
        return text_lower # Fallback if no keywords are found

df_modules['form_of_assessment'] = df_modules['form_of_assessment'].apply(clean_form_of_assessment)

print("Unique values in 'form_of_assessment' column after cleaning:")
display(df_modules['form_of_assessment'].value_counts())
```

```python
def clean_admission_requirements(text):
    text_str = str(text).strip().lower()
    if text_str in ['', 'none', 'nan', 'n/a']:
        return 'None'
    return text

df_modules['admission_requirements_for_assessment'] = df_modules['admission_requirements_for_assessment'].apply(clean_admission_requirements)

print("Unique values in 'admission_requirements_for_assessment' column after cleaning:")
display(df_modules['admission_requirements_for_assessment'].value_counts())
```

```python
def clean_duration_of_assessment(text):
    text_str = str(text).strip().lower()
    if text_str in ['', 'none', 'nan', 'n/a', 'continuous', 'varies']:
        return 'Not applicable'
    return text

df_modules['duration_of_assessment'] = df_modules['duration_of_assessment'].apply(clean_duration_of_assessment)

print("Unique values in 'duration_of_assessment' column after cleaning:")
display(df_modules['duration_of_assessment'].value_counts())
```

```python
def clean_offering(text):
    text_lower = str(text).lower()
    if 'fall semester' in text_lower and 'spring semester' in text_lower:
        return 'Both semesters'
    elif 'fall semester' in text_lower:
        return 'Fall semester'
    elif 'spring semester' in text_lower:
        return 'Spring semester'
    return text

df_modules['offering'] = df_modules['offering'].apply(clean_offering)

print("Unique values in 'offering' column after cleaning:")
display(df_modules['offering'].value_counts())
```

```python
import ast
import pandas as pd
import numpy as np
import re

def clean_list_column(entry):
    if entry is None or (isinstance(entry, float) and pd.isna(entry)):
        return []
    if isinstance(entry, str):
        stripped_entry = entry.strip()
        if stripped_entry == '' or stripped_entry.lower() == 'none' or stripped_entry.lower() == 'nan':
            return []
    if isinstance(entry, list) and not entry:
        return []

    items = []
    if isinstance(entry, str):
        # Try to parse as a list if it looks like one, e.g., "['module1', 'module2']"
        if stripped_entry.startswith('[') and stripped_entry.endswith(']') and stripped_entry != '[]':
            try:
                parsed_list = ast.literal_eval(stripped_entry)
                if isinstance(parsed_list, list):
                    items.extend(parsed_list)
                else:
                    items.append(str(parsed_list).strip())
            except (ValueError, SyntaxError):
                items.extend([item.strip() for item in stripped_entry.split(',')])
        else:
            items.extend([item.strip() for item in stripped_entry.split(',')])
    elif isinstance(entry, list):
        items.extend([str(item).strip() for item in entry])
    else:
        items.extend([item.strip() for item in str(entry).split(',')])

    cleaned_items = []
    seen = set()
    for item in items:
        if item and item not in seen:
            cleaned_items.append(item)
            seen.add(item)
    return cleaned_items

# Apply the cleaning to 'further_modules' to ensure it's always a list of strings
df_modules['further_modules'] = df_modules['further_modules'].apply(clean_list_column)

# Create a mapping for standardizing module names using moduleno and name
module_standard_names = {}
for index, row in df_modules.iterrows():
    mod_no = str(row['moduleno']).strip()
    mod_name = str(row['name']).strip()

    canonical_form = f"{mod_no} – {mod_name}"

    # Normalize name and module number for creating lookup keys
    normalized_mod_name_key_hyphenated = mod_name.lower().replace(' ', '-').replace('–', '-').replace('—', '-').strip()
    normalized_mod_no_key_hyphenated = mod_no.lower().replace(' ', '-').replace('–', '-').replace('—', '-').strip()

    # Add various lookup keys to the dictionary, all mapping to the same canonical_form
    module_standard_names[mod_no.lower()] = canonical_form # e.g., cs 651
    module_standard_names[normalized_mod_no_key_hyphenated] = canonical_form # e.g., cs-651

    module_standard_names[mod_name.lower()] = canonical_form # e.g., cryptography ii
    module_standard_names[normalized_mod_name_key_hyphenated] = canonical_form # e.g., cryptography-ii

    # Combined forms
    module_standard_names[f"{mod_no.lower()} – {mod_name.lower()}"] = canonical_form
    module_standard_names[f"{mod_no.lower()} - {mod_name.lower()}"] = canonical_form
    module_standard_names[f"{mod_no.lower()} {mod_name.lower()}"] = canonical_form
    module_standard_names[f"{normalized_mod_no_key_hyphenated}-{normalized_mod_name_key_hyphenated}"] = canonical_form
    module_standard_names[f"{normalized_mod_name_key_hyphenated}({normalized_mod_no_key_hyphenated})"] = canonical_form

    # Add common shorter forms/aliases from module names as keys
    # Content within parentheses
    match_parentheses = re.search(r'\((\s*[^)]+\s*)\)', mod_name, re.IGNORECASE)
    if match_parentheses:
        paren_content = match_parentheses.group(1).strip()
        module_standard_names[paren_content.lower()] = canonical_form
        module_standard_names[paren_content.lower().replace(' ', '-').replace('–', '-').replace('—', '-')] = canonical_form

    # Parts of name separated by dashes (e.g., 'Cryptography II' from 'CS 651 – Cryptography II')
    # This captures the part AFTER the first dash, assuming the number is before it.
    # Example: 'CS 651 – Cryptography II' -> 'Cryptography II'
    match_dash_split = re.split(r'\s*[-–]\s*', mod_name, maxsplit=1)
    if len(match_dash_split) > 1:
        part_after_dash = match_dash_split[1].strip()
        module_standard_names[part_after_dash.lower()] = canonical_form
        module_standard_names[part_after_dash.lower().replace(' ', '-').replace('–', '-').replace('—', '-')] = canonical_form

    # And the part before the first dash, if it's not just the number
    if len(match_dash_split) > 0:
        part_before_dash = match_dash_split[0].strip()
        # Only add if it's not already covered by the module number itself and seems like a meaningful name part
        if part_before_dash.lower() != mod_no.lower() and not re.match(r'^[A-Za-z]{2,}\s*\d{2,}$', part_before_dash):
            module_standard_names[part_before_dash.lower()] = canonical_form
            module_standard_names[part_before_dash.lower().replace(' ', '-').replace('–', '-').replace('—', '-')] = canonical_form



# Function to standardize items in further_modules list (more robust matching)
def standardize_further_modules_names(module_list):
    standardized_list = []
    for item in module_list:
        found_canonical = None

        # Prepare candidate keys for lookup based on the item
        candidate_lookup_keys = []

        # Attempt 1: Direct lowercase and stripped item
        lower_item = item.lower().strip()
        candidate_lookup_keys.append(lower_item)

        # Attempt 2: Item with spaces/dashes replaced by hyphens
        hyphenated_item = lower_item.replace(' ', '-').replace('–', '-').replace('—', '-')
        candidate_lookup_keys.append(hyphenated_item)

        # Attempt 3: If item has a module number prefix, try to extract just the name part
        # Example: 'CS 651 – Cryptography II' -> 'Cryptography II'
        # Use a regex to capture anything after a potential module code and separator
        # This handles cases like 'CS 651 – Name', 'IE 500 Name', 'CS500-Name'
        match_name_part = re.match(r'^(?:[A-Za-z]{2,}\s*\d{2,}[-–]?\s*)(.*)$', item.strip(), re.IGNORECASE)
        if match_name_part:
            name_part_only = match_name_part.group(1).strip() # This is the part after the potential module number
            if name_part_only:
                candidate_lookup_keys.append(name_part_only.lower()) # Lowercase name part
                candidate_lookup_keys.append(name_part_only.lower().replace(' ', '-').replace('–', '-').replace('—', '-')) # Hyphenated name part

        # Iterate through candidate keys to find a match in module_standard_names
        for key_to_check in candidate_lookup_keys:
            if key_to_check in module_standard_names:
                found_canonical = module_standard_names[key_to_check]
                break # Found a match, use it and stop checking candidate keys

        if found_canonical:
            standardized_list.append(found_canonical)
        else:
            standardized_list.append(item) # Keep original if no match (should be rare with robust mapping)

    # Remove duplicates after standardization and return
    return list(dict.fromkeys(standardized_list))


df_modules['further_modules'] = df_modules['further_modules'].apply(standardize_further_modules_names)

print("Unique individual modules in 'further_modules' column after cleaning and standardization:")
all_further_modules = []
for sublist in df_modules['further_modules']:
    all_further_modules.extend(sublist)
unique_further_modules_series = pd.Series(all_further_modules).value_counts().sort_index()
print(df_modules['further_modules'].value_counts())
```

```python
import re
import pandas as pd
import numpy as np

def extract_and_format_module_numbers(module_list):
    extracted_numbers = []
    for module_str in module_list:
        # Regex to find module numbers like "CS 500" or "IE 675b"
        # This assumes module numbers start with 2+ letters followed by optional space and digits/letters
        match = re.search(r'([A-Za-z]{2,}\s*\d+[a-zA-Z]?)', module_str)
        if match:
            extracted_numbers.append(match.group(1).strip())
        # Entries without a clear module number (e.g., 'Bachelor seminar...') are now ignored

    unique_numbers = list(dict.fromkeys(extracted_numbers)) # Ensure uniqueness of extracted numbers

    if unique_numbers:
        return ','.join(unique_numbers) # Join with comma, no spaces
    else:
        return np.nan # If no module numbers were extracted, return NaN

# Apply the new extraction and formatting function to the further_modules column
df_modules['further_modules'] = df_modules['further_modules'].apply(extract_and_format_module_numbers)

print("Unique individual module numbers in 'further_modules' column after extraction and formatting:")
# For this display, we need to handle the new string format.
# We can either split the strings back into lists or just display value_counts() on the column directly.
# Displaying value_counts() on the column will show unique comma-separated strings.
# If we want a flattened list of *all* individual module numbers across all rows, we need to process further.

# Let's display the value counts of the column as formatted strings first:
display(df_modules['further_modules'].value_counts(dropna=False))

print("\nExample of 'further_modules' column after formatting (first 5 rows):")
display(df_modules['further_modules'].head())
```

```python
import re
import numpy as np # Import numpy for np.nan

def clean_semester(text):
    # Handle initial NaN/None/empty strings by returning np.nan
    if pd.isna(text) or str(text).strip() == '':
        return np.nan

    text_lower = str(text).lower()

    # Normalize range formats to a simple space-separated list of numbers
    text_lower = text_lower.replace("2nd to 4th semester", "2 3 4")
    text_lower = text_lower.replace("2nd-4th semester", "2 3 4")
    text_lower = text_lower.replace("1st or 2nd semester", "1 2") # Explicitly handle 'or' for consistency
    text_lower = text_lower.replace("1st / 2nd semester", "1 2") # Explicitly handle '/' for consistency
    text_lower = text_lower.replace("1st/2nd/3rd semester", "1 2 3") # For modules like CS 605's master part
    text_lower = text_lower.replace("1st / 2nd semester", "1 2")
    text_lower = text_lower.replace("2nd/3rd semester", "2 3")
    text_lower = text_lower.replace("1st / 2nd semester", "1 2")

    # Remove all 'b.sc', 'm.sc', and 'semester' text, and ordinal suffixes
    # Replace non-digit characters (except space) with spaces
    text_lower = re.sub(r'\b(?:semester|b\.sc\.?|m\.sc\.?|st|nd|rd|th)\b', ' ', text_lower)
    text_lower = re.sub(r'[^\d]+', ' ', text_lower).strip()

    # Extract all potential semester numbers
    all_found_nums_str = re.findall(r'\d+', text_lower)

    master_semesters_found = set()
    for num_str in all_found_nums_str:
        try:
            num = int(num_str)
            # Keep only 1, 2, 3, 4 (Master's relevant)
            if 1 <= num <= 4:
                master_semesters_found.add(str(num))
        except ValueError:
            continue # Skip if it's not a valid number (e.g., from some other text)

    if master_semesters_found:
        # Sort for consistent output (e.g., '1,2,3')
        parts_sorted = sorted(list(master_semesters_found), key=int)
        return ','.join(parts_sorted)
    else:
        # If no relevant Master's semesters were found after filtering, return np.nan
        return np.nan

# Apply the cleaning directly to the 'semester' column
df_modules['semester'] = df_modules['semester'].apply(clean_semester)

# Drop the 'original_semester' column if it exists from previous runs
if 'original_semester' in df_modules.columns:
    df_modules.drop('original_semester', axis=1, inplace=True)

print("Unique values in 'semester' column after cleaning:")
display(df_modules['semester'].value_counts())

# Display the row(s) that are now np.nan (to confirm no unexpected NaN entries)
print("\nRow(s) where 'semester' is now NaN (to confirm no unexpected NaN entries):")
display(df_modules[df_modules['semester'].isna()])
```

```python
def standardize_range_of_application(applications_list):
    standardized_applications = []
    for app in applications_list:
        app_lower = str(app).lower()
        if 'mmds' in app_lower or 'mannheim master in data science' in app_lower:
            standardized_applications.append('M.Sc. Mannheim Master in Data Science')
        elif 'mmsds' in app_lower or 'mannheim master in social data science' in app_lower:
            standardized_applications.append('M.Sc. Mannheim Master in Social Data Science')
        elif 'm.sc. wirtschaftsinformatik' in app_lower or 'm.sc. business informatics' in app_lower:
            standardized_applications.append('M.Sc. Wirtschaftsinformatik')
        elif 'b.sc. wirtschaftsinformatik' in app_lower:
            standardized_applications.append('B.Sc. Wirtschaftsinformatik')
        elif 'lehramt informatik' in app_lower:
            standardized_applications.append('Lehramt Informatik')
        elif 'm.sc. mathematik' in app_lower:
            standardized_applications.append('M.Sc. Mathematik')
        elif 'm.sc. wirtschaftsmathematik' in app_lower:
            standardized_applications.append('M.Sc. Wirtschaftsmathematik')
        elif 'm.sc. wirtschaftspädagogik' in app_lower:
            standardized_applications.append('M.Sc. Wirtschaftspädagogik')
        elif 'minor applied computer science' in app_lower:
            standardized_applications.append('Minor Applied Computer Science')
        else:
            standardized_applications.append(app.strip()) # Keep original if not matched, but strip whitespace

    # Remove duplicates and return as a sorted list for consistency
    return sorted(list(set(standardized_applications)))


df_modules['range_of_application'] = df_modules['range_of_application'].apply(standardize_range_of_application)

print("Unique values in 'range_of_application' column after standardization:")
# To get value counts for individual items within lists, we need to flatten the list first
all_applications = []
for app_list in df_modules['range_of_application']:
    all_applications.extend(app_list)

display(pd.Series(all_applications).value_counts())
```
